```python
app_code = '''
from flask import Flask, request, jsonify, send_file
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, pipeline
import cv2
import whisper
from paddleocr import PaddleOCR
import subprocess
import os
import tempfile
import uuid
import threading
import numpy as np
import shutil

from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain.docstore.document import Document
from langchain.chains import RetrievalQA
from langchain_community.llms import HuggingFacePipeline
from langchain.prompts import PromptTemplate

app = Flask(__name__)
model = None
tokenizer = None
llm_pipeline = None
embeddings = None
ocr_reader = None
jobs = {}
FRAME_RATE = 150

GDRIVE_CHROMA_PATH = "/content/drive/MyDrive/lecture_rag/chromadb"
GDRIVE_VIDEO_PATH = "/content/drive/MyDrive/lecture_rag/videos"
GDRIVE_CLIP_PATH = "/content/drive/MyDrive/lecture_rag/clips"

def process_video_task(job_id, video_path):
    try:
        jobs[job_id]["status"] = "processing"
        jobs[job_id]["progress"] = 10
        jobs[job_id]["message"] = "Extracting audio..."

        audio_path = os.path.join(tempfile.gettempdir(), f"{job_id}.wav")
        subprocess.run(
            ["ffmpeg", "-i", video_path, "-vn", "-acodec", "pcm_s16le", "-ar", "16000", "-ac", "1", "-y", audio_path],
            stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, check=True
        )

        jobs[job_id]["progress"] = 25
        jobs[job_id]["message"] = "Transcribing with Whisper..."

        whisper_model = whisper.load_model("base")
        result = whisper_model.transcribe(audio_path, verbose=False)
        segments = result.get("segments", [])
        transcript = [{"start": s["start"], "end": s["end"], "text": s.get("text", "").strip()} for s in segments]
        duration = segments[-1]["end"] if segments else 0

        jobs[job_id]["progress"] = 50
        jobs[job_id]["message"] = "Extracting frames with OCR..."

        cap = cv2.VideoCapture(video_path)
        fps = cap.get(cv2.CAP_PROP_FPS)
        frames_data = []
        frame_count = 0
        saved = 0
        all_ocr_text = []

        while True:
            ret, frame = cap.read()
            if not ret:
                break
            if frame_count % FRAME_RATE == 0:
                timestamp = frame_count / fps
                results = ocr_reader.ocr(frame, cls=True)
                printed_text = []
                if results and results[0]:
                    for line in results[0]:
                        text = line[1][0]
                        conf = line[1][1]
                        if conf >= 0.5:
                            printed_text.append(text)
                joined = " ".join(printed_text)
                all_ocr_text.append(joined)
                frames_data.append({
                    "timestamp": timestamp,
                    "printed_text": joined,
                    "handwritten_text": "",
                    "ocr_confidence": 0.0
                })
                saved += 1
            frame_count += 1
        cap.release()

        jobs[job_id]["progress"] = 80
        jobs[job_id]["message"] = "Building LangChain RAG index with timestamps..."

        # Build documents with timestamps from Whisper segments
        documents = []
        for seg in transcript:
            if seg["text"].strip():
                documents.append(Document(
                    page_content=seg["text"],
                    metadata={
                        "start_time": seg["start"],
                        "end_time": seg["end"],
                        "source": "whisper",
                        "job_id": job_id
                    }
                ))

        # Add OCR text with timestamps
        for frame in frames_data:
            if frame["printed_text"].strip():
                documents.append(Document(
                    page_content=frame["printed_text"],
                    metadata={
                        "start_time": max(0, frame["timestamp"] - 2),
                        "end_time": frame["timestamp"] + 2,
                        "source": "ocr",
                        "job_id": job_id
                    }
                ))

        # LangChain smart chunking
        splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50)
        split_docs = splitter.split_documents(documents)

        # Store in ChromaDB via LangChain
        collection_path = os.path.join(GDRIVE_CHROMA_PATH, f"job_{job_id}")
        os.makedirs(collection_path, exist_ok=True)
        vectorstore = Chroma.from_documents(
            documents=split_docs,
            embedding=embeddings,
            persist_directory=collection_path,
            collection_name=f"job_{job_id}"
        )

        # Save video permanently to Google Drive
        os.makedirs(GDRIVE_VIDEO_PATH, exist_ok=True)
        saved_video_path = os.path.join(GDRIVE_VIDEO_PATH, f"{job_id}.mp4")
        shutil.copy2(video_path, saved_video_path)

        jobs[job_id]["progress"] = 95
        jobs[job_id]["message"] = "Finalizing..."
        jobs[job_id]["result"] = {
            "transcript": transcript,
            "frames": frames_data,
            "duration": duration,
            "chunks_indexed": len(split_docs)
        }
        jobs[job_id]["status"] = "completed"
        jobs[job_id]["progress"] = 100
        jobs[job_id]["message"] = "Done"

        os.remove(video_path)
        if os.path.exists(audio_path):
            os.remove(audio_path)

    except Exception as e:
        jobs[job_id]["status"] = "failed"
        jobs[job_id]["error"] = str(e)
        jobs[job_id]["message"] = str(e)

@app.route("/health", methods=["GET"])
def health_check():
    return jsonify({"status": "healthy", "model_loaded": llm_pipeline is not None}), 200

@app.route("/upload", methods=["POST"])
def upload_video():
    try:
        if "video" not in request.files:
            return jsonify({"error": "No video file"}), 400
        file = request.files["video"]
        if file.filename == "":
            return jsonify({"error": "No file selected"}), 400
        job_id = str(uuid.uuid4())
        video_path = os.path.join(tempfile.gettempdir(), f"{job_id}.mp4")
        file.save(video_path)
        jobs[job_id] = {"status": "processing", "progress": 0, "message": "Starting...", "result": None, "error": None}
        threading.Thread(target=process_video_task, args=(job_id, video_path)).start()
        return jsonify({"job_id": job_id}), 200
    except Exception as e:
        return jsonify({"error": str(e)}), 500

@app.route("/status/<job_id>", methods=["GET"])
def get_status(job_id):
    if job_id not in jobs:
        return jsonify({"error": "Job not found"}), 404
    j = jobs[job_id]
    return jsonify({"status": j["status"], "progress": j["progress"], "message": j["message"], "error": j.get("error")}), 200

@app.route("/result/<job_id>", methods=["GET"])
def get_result(job_id):
    if job_id not in jobs:
        return jsonify({"error": "Job not found"}), 404
    if jobs[job_id]["status"] != "completed":
        return jsonify({"error": "Job not ready"}), 400
    return jsonify(jobs[job_id]["result"]), 200

@app.route("/collections", methods=["GET"])
def list_collections():
    try:
        cols = []
        if os.path.exists(GDRIVE_CHROMA_PATH):
            for name in os.listdir(GDRIVE_CHROMA_PATH):
                if name.startswith("job_"):
                    cols.append(name.replace("job_", ""))
        return jsonify({"collections": cols}), 200
    except Exception as e:
        return jsonify({"error": str(e)}), 500

@app.route("/job/<job_id>", methods=["DELETE"])
def delete_job(job_id):
    try:
        collection_path = os.path.join(GDRIVE_CHROMA_PATH, f"job_{job_id}")
        if os.path.exists(collection_path):
            shutil.rmtree(collection_path)
        video_path = os.path.join(GDRIVE_VIDEO_PATH, f"{job_id}.mp4")
        if os.path.exists(video_path):
            os.remove(video_path)
        if job_id in jobs:
            del jobs[job_id]
        return jsonify({"ok": True}), 200
    except Exception as e:
        return jsonify({"error": str(e)}), 500

@app.route("/generate", methods=["POST"])
def generate():
    try:
        data = request.json
        prompt = data.get("prompt", "")
        job_id = data.get("job_id", "")
        max_tokens = data.get("max_tokens", 500)
        temperature = data.get("temperature", 0.7)

        # Casual message detection
        casual_triggers = ["hi", "hello", "hey", "how are you", "thanks", "thank you", "bye", "ok", "okay"]
        is_casual = prompt.strip().lower() in casual_triggers or len(prompt.strip().split()) <= 2

        if is_casual:
            pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=100)
            response = pipe(f"<s>[INST] Reply naturally to: {prompt} [/INST]")[0]["generated_text"]
            return jsonify({"text": response.split("[/INST]")[-1].strip(), "metadata": {"rag_used": False}}), 200

        if not job_id:
            return jsonify({"error": "job_id is required for lecture questions"}), 400

        # Load LangChain vectorstore for this job
        collection_path = os.path.join(GDRIVE_CHROMA_PATH, f"job_{job_id}")
        if not os.path.exists(collection_path):
            return jsonify({"error": "Lecture not found. Please upload and process the video first."}), 404

        vectorstore = Chroma(
            persist_directory=collection_path,
            embedding_function=embeddings,
            collection_name=f"job_{job_id}"
        )

        retriever = vectorstore.as_retriever(search_kwargs={"k": 5})

        # Get relevant docs with metadata
        relevant_docs = retriever.get_relevant_documents(prompt)

        # Extract timestamps
        start_times = [doc.metadata.get("start_time", 0) for doc in relevant_docs]
        end_times = [doc.metadata.get("end_time", 0) for doc in relevant_docs]
        clip_start = max(0, min(start_times) - 2)
        clip_end = max(end_times) + 2

        context = "\n\n".join([doc.page_content for doc in relevant_docs])
        if len(context) > 6000:
            context = context[:6000] + "\n\n[Context truncated...]"

        full_prompt = f"""<s>[INST] Based on the following lecture content, answer the question.

Lecture Content:
{context}

Question: {prompt} [/INST]"""

        inputs = tokenizer(full_prompt, return_tensors="pt", truncation=True, max_length=4096).to(model.device)

        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=max_tokens,
                temperature=temperature,
                do_sample=True,
                pad_token_id=tokenizer.eos_token_id
            )

        prompt_len = inputs["input_ids"].shape[1]
        generated_ids = outputs[0][prompt_len:]
        answer = tokenizer.decode(generated_ids, skip_special_tokens=True).strip()

        # Cut video clip using ffmpeg
        clip_id = str(uuid.uuid4())
        os.makedirs(GDRIVE_CLIP_PATH, exist_ok=True)
        clip_path = os.path.join(GDRIVE_CLIP_PATH, f"{clip_id}.mp4")
        video_source = os.path.join(GDRIVE_VIDEO_PATH, f"{job_id}.mp4")

        clip_available = False
        if os.path.exists(video_source):
            try:
                subprocess.run([
                    "ffmpeg", "-i", video_source,
                    "-ss", str(clip_start),
                    "-to", str(clip_end),
                    "-c", "copy", "-y", clip_path
                ], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, check=True)
                clip_available = True
            except Exception:
                clip_available = False

        return jsonify({
            "text": answer,
            "clip_id": clip_id if clip_available else None,
            "timestamp": {
                "start": clip_start,
                "end": clip_end,
                "label": f"{int(clip_start//60)}:{int(clip_start%60):02d} - {int(clip_end//60)}:{int(clip_end%60):02d}"
            },
            "metadata": {
                "prompt_length": len(full_prompt),
                "tokens_generated": len(generated_ids),
                "rag_used": True,
                "chunks_used": len(relevant_docs)
            }
        }), 200

    except Exception as e:
        return jsonify({"error": str(e)}), 500

@app.route("/clip/<clip_id>", methods=["GET"])
def get_clip(clip_id):
    clip_path = os.path.join(GDRIVE_CLIP_PATH, f"{clip_id}.mp4")
    if not os.path.exists(clip_path):
        return jsonify({"error": "Clip not found"}), 404
    return send_file(clip_path, mimetype="video/mp4")

def load_llm():
    global model, tokenizer, llm_pipeline, embeddings, ocr_reader

    print("Loading PaddleOCR...")
    ocr_reader = PaddleOCR(use_angle_cls=True, lang="en", use_gpu=False, show_log=False)
    print("PaddleOCR ready.")

    print("Loading LangChain embeddings...")
    embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
    print("Embeddings ready.")

    os.makedirs(GDRIVE_CHROMA_PATH, exist_ok=True)
    os.makedirs(GDRIVE_VIDEO_PATH, exist_ok=True)
    os.makedirs(GDRIVE_CLIP_PATH, exist_ok=True)
    print(f"Drive folders ready.")

    model_name = "mistralai/Mistral-7B-Instruct-v0.2"
    print(f"Loading model: {model_name}")

    tokenizer = AutoTokenizer.from_pretrained(model_name)
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token

    quant_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_compute_dtype=torch.float16,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_use_double_quant=True
    )

    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        quantization_config=quant_config,
        device_map="auto"
    )

    print("Mistral 7B loaded successfully!")

if __name__ == "__main__":
    load_llm()
    app.run(host="0.0.0.0", port=5000)
'''

with open("app.py", "w") as f:
    f.write(app_code)

print("app.py written!")
```
